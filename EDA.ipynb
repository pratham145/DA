```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking the null values in each column
df.isnull().sum()
```

```python
#handling the missing values 
# dropping rows with missing values
df.dropna(inplace=True)

```

```python
#checking if missing values removed or not
df.isnull().sum()

```

```python
#exploring datatypes and summary statistics
df.dtypes
```

```python
df.describe()
```

```python
#Visualizing the data

#histogram
df.hist(figsize=(10,10))
```

```python
# Scatter plot matrix
plt.figure(figsize=(8, 6))
sns.scatterplot(x='infant deaths', y='Year', data=df)
plt.title('Infant Deaths V/s Years')
plt.xlabel('infant Deaths')
plt.ylabel('Years')
plt.show()
```

```python
#count plot 
sns.countplot(x='infant deaths', data=df)
```

```python
# Box PLOT
sns.boxplot(x='infant deaths', y='Year', data=df)
```
