```python
import pandas as pd

# Load the data
file_path = 'Life Expectancy Data.csv'  # Replace this with the actual file path if it's different

# Try reading the file with different encodings
encodings = ['latin1', 'iso-8859-1', 'cp1252']
for encoding in encodings:
    try:
        df = pd.read_csv(file_path, encoding=encoding)
        print(f"Successfully read the file with '{encoding}' encoding")
        break
    except UnicodeDecodeError:
        print(f"Failed to read the file with '{encoding}' encoding")
else:
    print("Could not read the file with any of the tried encodings")

# Display the first few rows of the dataframe if successful
if 'df' in locals():
    print(df.head())
else:
    print("Could not read the file with any of the tried encodings")
```
